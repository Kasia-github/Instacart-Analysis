# 03. Wrangling data

```python
# Drop First Name and Surnam column
df_customers = df_customers.drop(columns = ['First Name', 'Surnam'])
```

```python
df_customers.head()
```

```python
# Rename columns
df_customers = df_customers.rename(columns={'Gender': 'gender', 'STATE': 'state', 'Age': 'age', 'n_dependants': 'children'})
```

```python
df_customers.head()
```

```python
# Replace values in 'gender' column
df_customers['gender'] = df_customers['gender'].replace({'Male': 'm', 'Female': 'f'})
```

```python
df_customers.head()
```

```python
# Replace values in 'fam_status' column
df_customers['fam_status'] = df_customers['fam_status'].replace({'married': 'm', 'single': 's'})
```

```python
df_customers.head()
```

```python
df_customers.shape
```

```python
df_customers.info()
```

```python
# Define a dictionary with the columns to change and their new data types
dtype_dict = {
    'user_id': 'int32',
    'age': 'int32',
    'children': 'int32',
    'income': 'int32'}
```

```python
# Use the astype method with the dictionary to change the data types
df_customers = df_customers.astype(dtype_dict)
```

```python
# Verify the changes
print(df_customers.dtypes)
```

```python
# Check for mixed type columns
for col in df_customers.columns.tolist():
  weird = (df_customers[[col]].applymap(type) != df_customers[[col]].iloc[0].apply(type)).any(axis = 1)
if len (df_customers[weird]) > 0: print (col)
```

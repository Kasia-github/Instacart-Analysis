```python
# Define a function
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
# Apply the function
df['price_range'] = df.apply(price_label, axis=1)
```

```python
# Create a new column busiest_period_of_day' with If-Statement with User-Defined Function
```

```python
df_ords_prods_merge['order_hour_of_day'].value_counts(dropna = False)
```

```python
# Define a function
def busiest_time(row):

  if row['order_hour_of_day'] >= 2454203:
    return 'Most orders'
  elif (row['order_hour_of_day'] > 402316) and (row['order_hour_of_day'] < 2454203):
    return 'Average orders'
  elif row['order_hour_of_day'] <= 402316:
    return 'Fewest orders'
  else: return 'Not enough data'
```

```python
# Apply the function
df_ords_prods_merge['busiest_period_of_day'] = df_ords_prods_merge.apply(busiest_time, axis=1)
```

```python
df_ords_prods_merge['busiest_period_of_day'].value_counts(dropna = False)
```

```python
# 2. Option -> Create a new column 'busiest_period_of_day' with If-Statement with for-loops
```

```python
result = []

for value in df_ords_prods_merge["order_hour_of_day"]:
  if value in [10, 11, 14, 15, 13, 12, 16, 9]:
    result.append("Most orders")
  elif value in [23, 6, 0, 1, 5, 2, 4, 3]:
    result.append("Fewest orders")
  else:
    result.append("Average orders")
```

```python
result
```

```python
# Derive new column "busiest_period_of_day"

df_ords_prods_merge['busiest_period_of_day'] = result
```

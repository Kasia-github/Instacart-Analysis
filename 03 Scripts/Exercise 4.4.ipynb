```python
# Import data
df_ords = pd.read_csv(r'C:\Users\kasia\Desktop\Career Foundry\Achievement 4\06-2024 Instacart Basket Analysis\02 Data\Original Data\orders.csv', index_col = False)
```

```python
df_ords.head()
```

```python
# Droppping column eval_set from orders.csv
df_ords.drop(columns = ['eval_set'])
```

```python
#  Renaming the user_id column
df_ords.rename(columns = {'user_id' : 'customer_id'}, inplace = True)
```

```python
df_ords.head()
```

```python
#  Renaming the order_dow column
df_ords.rename(columns = {'order_dow' : 'orders_day_of_week'}, inplace = True)
```

```python
df_ords.head()
```

```python
df_ords[df_ords['customer_id'] == 1]
```

```python
user_info = df_ords[df_ords['customer_id'] == 1]
```

```python
print(user_info)
```

```python
# Total number of orders
total_orders = user_info['order_id'].nunique()
```

```python
# Order frequency by day of the week and hour of the day
orders_by_day_of_week = user_info['orders_day_of_week'].value_counts()
order_hour_of_day = user_info['order_hour_of_day'].value_counts()
```

```python
# Average days since prior order
average_days_since_prior = user_info['days_since_prior_order'].mean()
```

```python
# First and last order numbers
first_order_number = user_info['order_number'].min()
last_order_number = user_info['order_number'].max()
```

```python
# Total number of days active (considering 'days_since_prior_order')
# Sum of days since prior order, ignoring NaN for the first order
total_days_active = user_info['days_since_prior_order'].sum()
```

```python
# Adding 0 for the first order
if np.isnan(user_info['days_since_prior_order'].iloc[0]):
    total_days_active += 0
```

```python
# Print the summary statistics
print(f"Customer ID: 1")
print(f"Total Number of Orders: {total_orders}")
print(f"Orders by Day of the Week:\n{orders_by_day_of_week}")
print(f"Orders by Hour of the Day:\n{order_hour_of_day}")
print(f"Average Days Since Prior Order: {average_days_since_prior:.2f}")
print(f"First Order Number: {first_order_number}")
print(f"Last Order Number: {last_order_number}")
print(f"Total Number of Days Active: {total_days_active}")
```

```python
# Create an 'average_price' column
df_ords_prods_merge['average_price'] = df_ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
# Create a spend flag using loc
df_ords_prods_merge.loc[df_ords_prods_merge['average_price']<10,'spend_flag']='Low_spender'

df_ords_prods_merge.loc[df_ords_prods_merge['average_price']>= 10, 'spend_flag'] = 'High_spender'
```

```python
# Check spend flag
df_ords_prods_merge['spend_flag'].value_counts(dropna = False)
```

```python
df_ords_prods_merge[['user_id', 'average_price', 'spend_flag', 'order_number' , 'prices']].head(60)
```

```python
# Create a frequent_customer column
df_ords_prods_merge['frequent_customer'] = df_ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
# Create order_frequency_flag
df_ords_prods_merge.loc[df_ords_prods_merge['frequent_customer'] > 20, 'order_frequency_flag'] = 'Non_frequent_customer'

df_ords_prods_merge.loc[(df_ords_prods_merge['frequent_customer']>10) & (df_ords_prods_merge['frequent_customer']<=20),'order_frequency_flag'] = 'Regular customer'

df_ords_prods_merge.loc[df_ords_prods_merge['frequent_customer'] <= 10, 'order_frequency_flag'] = 'Frequent Customer'
```

```python
# Check order_frequency_flag

df_ords_prods_merge['order_frequency_flag'].value_counts(dropna = False)
```
